```python
filename = 'informe_acciones.txt'
with open(filename, 'r', encoding='utf-8', errors='replace') as file:
    data = file.readlines()

import re
import numpy as np

def extract_data(lines):
    company_pattern = r"Asunto: (.*?) - Informe"
    date_pattern = r"\b(\d{4}-\d{2}-\d{2})\b"
    value_pattern = r"Valor actual: ([\d\.]+)"

    companies = {}

    for line in lines:
        company_match = re.search(company_pattern, line)
        date_match = re.search(date_pattern, line)
        value_match = re.search(value_pattern, line)

        if company_match and date_match and value_match:
            company = company_match.group(1)
            date = date_match.group(1)
            value = float(value_match.group(1))

            if company not in companies:
                companies[company] = []

            companies[company].append((date, value))

    return companies

companies_data = extract_data(data)

def analyze_companies(companies):
    results = []

    for company, records in companies.items():
        records.sort(reverse=True, key=lambda x: x[0])
        dates, values = zip(*records)

        latest_date = dates[0]
        max_value = max(values)
        min_value = min(values)
        mean_value = np.mean(values)
        variance = np.var(values)
        std_dev = np.std(values)

        results.append([company, latest_date, max_value, min_value, mean_value, variance, std_dev])

    return results

results = analyze_companies(companies_data)

import pandas as pd

df = pd.DataFrame(results, columns=["Empresa", "Fecha más reciente", "Valor Máx ($)", "Valor Mín ($)", "Promedio ($)", "Varianza ($^2$)", "Std. Dev ($)"])
display(df)
```
