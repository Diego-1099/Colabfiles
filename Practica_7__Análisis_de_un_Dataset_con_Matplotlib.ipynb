# **Análisis de un Dataset con Matplotlib**

```python
import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset('iris')

print(df.head())
print(df.info())
```

a) Gráfico de barras (Comparar el promedio de largo y ancho de los pétalos entre las tres especies de iris)

```python
promedio_valores = df.groupby('species')[['petal_length', 'petal_width']].mean()

promedio_valores.plot(kind = 'bar', figsize = (10, 6), color = ['black', 'red'])
plt.title('Promedio de largo y ancho de pétalos por especie')
plt.xlabel('Especie')
plt.ylabel('Promedio')
plt.show()
```

b) Histograma (Distribución del largo de los sépalos)

```python
plt.Figure = plt.figure(figsize = (10, 6))

plt.hist(df['sepal_length'], bins = 15, color = 'red', edgecolor = 'black')
plt.title('Distribución del largo de los sépalos')
plt.xlabel('Largo del sépalo')
plt.ylabel('Frecuencia')
plt.show()
```

c) Gráfica de dispersión (Relación entre el largo y el ancho del pétalo)

```python
plt.Figure = plt.figure(figsize = (10, 6))

sns.scatterplot(data = df, x = 'petal_length', y = 'petal_width', hue = 'species', palette = 'Set1')
plt.title('Relación entre el largo y el ancho del pétalo')
plt.xlabel('Largo del pétalo')
plt.ylabel('Ancho del pétalo')
plt.legend(title = 'Especie')
plt.show()
```

d) Gráfica de cajas (Distribución de las medidas por sépalo por especie)

```python
plt.Figure = plt.figure(figsize = (10, 6))

sns.boxplot(data = df, x = 'species', y = 'sepal_length', hue = 'species', palette = 'Set1')
plt.title('Distribución de las medidas por sépalo por especie')
plt.xlabel('Especie')
plt.ylabel('Largo del sépalo')
plt.show()
```

e) Gráfico de líneas (Tendencia de la longitud del pétalo a través de las observaciones)

```python
plt.Figure = plt.figure(figsize = (10, 6))

plt.plot(df['petal_length'], color = 'black', linestyle = '-', marker = 'o')
plt.title('Tendencia de la longitud del pétalo a través de las observaciones')
plt.xlabel('Indice de las observaciones')
plt.ylabel('Largo del pétalo')
plt.show()
```
