# Análisis de un Dataset con Seaborn y visualización **avanzada**

```python
import seaborn as sns
import pandas as pd

df = sns.load_dataset('tips')

print(df.head())
print(df.info())
print(df.describe())
print(df.isnull().sum())
```

a) Heatmap (Visualizar la correlación entre las variables númericas en el dataset)

```python
import matplotlib.pyplot as plt

cor = df[['total_bill', 'tip', 'size']].corr()

plt.Figure(figsize = (8, 6))
sns.heatmap(cor, annot = True, cmap = 'coolwarm', fmt = '.2f', cbar = True)
plt.title('Correlación entre las variables númericas')
plt.show()
```

b) Diagrama de Voilín (Muestra la distribución de las propinas para diferentes días de la semana)

```python
plt.Figure(figsize = (8, 6))
sns.violinplot(data = df, x = 'day', y = 'tip', hue = 'day', palette = 'Set1')
plt.title('Distribución de las propinas por día de la semana')
plt.xlabel('Día de la semana')
plt.ylabel('Propina ($)')
plt.show()
```

c) Gráfico de dispersión (Analizar la relación entre el total de la cuenta y la propina)

```python
plt.Figure(figsize = (8, 6))
sns.scatterplot(data = df, x = 'total_bill', y = 'tip', hue = 'day', palette = 'Set1')
plt.title('Relación entre el total de la cuenta y la propina')
plt.xlabel('Total de la cuenta ($)')
plt.ylabel('Propina ($)')
plt.legend(title = 'Día de la semana')
plt.show()
```

d) Gráfica de barras (Comparar el promedio de la propina por género)

```python
plt.Figure(figsize = (8, 6))

sns.barplot(data = df, x = 'sex', y = 'tip', hue = 'sex', palette = 'Set1')
plt.title('Promedio de las propinas por género')
plt.xlabel('Género')
plt.ylabel('Promedio de la propina ($)')
plt.show()
```
